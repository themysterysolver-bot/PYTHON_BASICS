```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
print(matplotlib.__version__)
```

```python
xpoints=np.array([0,5])
ypoints=np.array([5,25])
print(xpoints,ypoints)
plt.plot(xpoints,ypoints)
plt.show()
#plotting (0,5) and (5,25)
```

```python
plt.plot(xpoints,ypoints,'o')  #rings
```

- We should have the same number of points.
- Default x-axis is 0,1,2,3

```python
plt.plot(np.array([1,3,7,9]),np.array([8,5,-7,2]),marker='*')
```

### Marker Reference
You can choose any of these markers:

Marker	Description
'o'	Circle
'*'	Star
'.'	Point
','	Pixel
'x'	X
'X'	X (filled)
'+'	Plus
'P'	Plus (filled)
's'	Square
'D'	Diamond
'd'	Diamond (thin)
'p'	Pentagon
'H'	Hexagon
'h'	Hexagon
'v'	Triangle Down
'^'	Triangle Up
'<'	Triangle Left
'>'	Triangle Right
'1'	Tri Down
'2'	Tri Up
'3'	Tri Left
'4'	Tri Right
'|'	Vline
'_'	Hline

---
### Line Reference
Line Syntax	Description
'-'	Solid line
':'	Dotted line
'--'	Dashed line
'-.'	Dashed/dotted line

---

### Color Reference
Color Syntax	Description
'r'	Red
'g'	Green
'b'	Blue
'c'	Cyan
'm'	Magenta
'y'	Yellow
'k'	Black
'w'	White

---
 - markersize or the shorter version, `ms` to set the size of the markers.
 - `marker|line|color`
 - mec-> marker edge color
 - mfc->marker face color
 - keyword argument `linestyle`, or shorter `ls`, to change the style of the plotted line.
 - keyword argument `color` or the shorter `c` to set the color of the line.
 - keyword argument `linewidth` or the shorter `lw` to change the width of the line.
 ---

```python
plt.plot(np.array([i for i in range(5)]),np.array([i for i in range(0,10,2)]),'h:b',ms=10,mec='r',mfc='g')
```

```python
x1=np.array([1,2,3,4])
y1=np.array([4,5,6,7])
x2=np.array([-1,-2,-3,-4])
y2=np.array([0,1,2,3])
plt.plot(x1,y2,x2,y2)
font={'family':'serif','color':'darkblue','size':15}
plt.xlabel('Distance',fontdict=font)
plt.ylabel('Time',fontdict=font)
plt.title('Distance Vs Time Graph',loc='left')
plt.grid() #axis='x' or axis='y' color,linestyle,linewidwth can be given to grid too!
```
