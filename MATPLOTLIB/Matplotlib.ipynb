```python
x1=np.array([1,2,3,4])
y1=np.array([4,5,6,7])
x2=np.array([-1,-2,-3,-4])
y2=np.array([0,1,2,3])
plt.plot(x1,y2,x2,y2)
font={'family':'serif','color':'darkblue','size':15}
plt.xlabel('Distance',fontdict=font)
plt.ylabel('Time',fontdict=font)
plt.title('Distance Vs Time Graph',loc='left')
plt.grid() #axis='x' or axis='y' color,linestyle,linewidwth can be given to grid too!
```

## SubPlots
- With the `subplot()` function you can draw multiple plots in one figure.
- The subplot() Function  takes three arguments that describes the layout of the figure.
- The layout is organized in ***rows*** and ***columns***, which are represented by the *first and second* argument.
- The *third* argument represents the ***index*** of the current plot.(1-based indexing)
- To add title to the entire figure with the `suptitle()` function:

```python
x3=np.array([i for i in range(10)])
y3=np.array([i for i in range(0,20,2)])
plt.plot(x3,y3)
x4=np.array([i for i in range(10)])
y4=np.array([i for i in range(10,20)])
plt.plot(x4,y4)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x3,y3)
plt.title('Graph-1')

plt.subplot(1,2,2)
plt.plot(x4,y4)
plt.title('Graph-2')

plt.suptitle('SubPlot Graphs!')
plt.show()
```

- The `scatter()` function plots one dot for each observation. It needs two arrays of the same length,

```python
colr=np.array(["red","yellow","blue","brown","cyan","red","yellow","blue","brown","cyan"])
plt.scatter(x3,y3,c=colr)
```

- you can set your own color for each scatter plot with the `color` or the `c` argument.
- can even set a *specific color* for **each dot** by using an ***array of colors*** as value for the c argument.
